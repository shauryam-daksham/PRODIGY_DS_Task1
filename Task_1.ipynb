```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df2= pd.read_csv('/content/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_448447.csv')
df2.head()
```

```python

region_counts = df2.groupby('IncomeGroup').size()


plt.figure(figsize=(10, 6))
region_counts.plot(kind='barh', color=sns.color_palette('Dark2'))


plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.title('Count of IncomeGroup')
plt.xlabel('Count')
plt.ylabel('IncomeGroup')
plt.show()
```

```python

# Group by 'Region' and count the occurrences
region_counts = df2.groupby('Region').size()

# Plot the horizontal bar plot
plt.figure(figsize=(10, 6))
region_counts.plot(kind='barh', color=sns.color_palette('Dark2'))

# Customize the plot to remove the top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)



plt.title('Count of Regions')
plt.xlabel('Count')
plt.ylabel('Region')
plt.show()
```

```python
counts = df2.groupby(['Region', 'IncomeGroup']).size().unstack().fillna(0)

# Plotting function for Region vs IncomeGroup bar plot
def plot_region_vs_income_bar():
    counts.plot(kind='bar', stacked=False, figsize=(12, 8))
    plt.title('Region vs Income Group - Bar Plot')
    plt.xlabel('Region')
    plt.ylabel('Count')
    plt.legend(title='Income Group')
    plt.show()

# Plotting function for Region vs IncomeGroup histogram (stacked bar plot)
def plot_region_vs_income_histogram():
    counts.plot(kind='bar', stacked=True, figsize=(12, 8))
    plt.title('Region vs Income Group - Stacked Bar Plot')
    plt.xlabel('Region')
    plt.ylabel('Count')
    plt.legend(title='Income Group')
    plt.show()

# Plot the Region vs IncomeGroup bar plot
plot_region_vs_income_bar()

# Plot the Region vs IncomeGroup stacked bar plot
plot_region_vs_income_histogram()
```
